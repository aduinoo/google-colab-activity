````python
import pandas as pd

# Load the CSV you just collected
df = pd.read_csv("/content/tweets-data/deepavali_celebration.csv")

# Preview a few rows
df.head()
````

````python
import re
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()                                  # lowercase
    text = re.sub(r"http\\S+|www\\S+", "", text)         # remove URLs
    text = re.sub(r"@\\w+", "", text)                    # remove mentions
    text = re.sub(r"[^a-z\\s#]", " ", text)              # remove punctuation/numbers
    text = re.sub(r"\\s+", " ", text).strip()            # collapse spaces
    return text

df["clean_text"] = df["full_text"].apply(clean_text)
df.head()
#clean and normalize text
````

````python
import spacy
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])

def tokenize_and_lemmatize(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if token.lemma_ not in stop_words and len(token) > 2]
    return " ".join(tokens)

df["processed_text"] = df["clean_text"].apply(tokenize_and_lemmatize)
df.head()
#tokenize,remove stopwords and lemmetize
````

````python
df.to_csv("deepavali_preprocessed.csv", index=False)
print("✅ Preprocessed data saved as deepavali_preprocessed.csv")
````

````python
!pip install wordcloud matplotlib seaborn
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Tokenize processed text
all_words = " ".join(df["processed_text"]).split()
word_freq = Counter(all_words)

# Top 20 words
top_words = word_freq.most_common(20)
words, counts = zip(*top_words)

# Bar chart
plt.figure(figsize=(10,6))
sns.barplot(x=list(counts), y=list(words))
plt.title("Top 20 Most Common Words about Deepavali in Malaysia")
plt.xlabel("Frequency")
plt.tight_layout()
plt.show()

# Word Cloud
wc = WordCloud(width=1000, height=500, background_color="white").generate(" ".join(all_words))
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud — Deepavali Discussions in Malaysia")
plt.show()
````

````python
!pip install vaderSentiment
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    scores = analyzer.polarity_scores(text)
    if scores['compound'] >= 0.05:
        return "Positive"
    elif scores['compound'] <= -0.05:
        return "Negative"
    else:
        return "Neutral"

df["sentiment"] = df["clean_text"].apply(get_sentiment)

# Plot sentiment distribution
sns.countplot(x="sentiment", data=df)
plt.title("Sentiment Distribution of Deepavali Tweets (Malaysia)")
plt.xlabel("Sentiment Category")
plt.ylabel("Count")
plt.show()

df["sentiment"].value_counts(normalize=True) * 100
````

````python
!pip install gensim
from gensim import corpora, models

# Prepare tokenized data
tokenized = [t.split() for t in df["processed_text"]]
dictionary = corpora.Dictionary(tokenized)
corpus = [dictionary.doc2bow(text) for text in tokenized]

# Train LDA model
lda = models.LdaModel(corpus, num_topics=4, id2word=dictionary, passes=10, random_state=42)

for i, topic in lda.print_topics():
    print(f"Topic {i+1}: {topic}\\n")
````

````python
import spacy
# Load the English language model for spaCy
nlp = spacy.load("en_core_web_sm")

# Function to extract named entities from text
def extract_entities(text):
    # Process the text with the spaCy model
    doc = nlp(text)
    # Return a list of tuples containing the entity text and its label
    return [(ent.text, ent.label_) for ent in doc.ents]

# Apply the entity extraction function to the 'clean_text' column and store results in a new column 'entities'
df["entities"] = df["clean_text"].apply(extract_entities)

# Flatten the list of entities and count their occurrences
from collections import Counter
entities_flat = [f"{t}|{l}" for ents in df["entities"] for (t, l) in ents]
# Print the 20 most common entities
Counter(entities_flat).most_common(20)
````

Conclusion

The Natural Language Processing (NLP) analysis of 200 Deepavali-related tweets in Malaysia provides a clear view of how Malaysians express and discuss the festival online.

1.  The sentiment analysis reveals that the majority of tweets are either neutral (66.87%) or positive (30.72%), indicating that Deepavali is largely viewed with positivity and celebration. The low percentage of negative sentiment (2.41%) suggests minimal controversy or negativity around the event.
2.   The word frequency analysis highlights a strong Malaysian cultural context, with many Malay words (e.g., hari, dan, cahaya, perayaan, menyambut) mixed with English terms. This linguistic blend reflects Malaysia’s multilingual society, where English and Malay are often used interchangeably in online discourse. The frequent presence of “https” indicates that users actively share links, news, or visual content (e.g., greetings, events, or promotions) related to Deepavali.



The topic modeling (LDA) uncovered four main themes:

1. Social Media and Online Sharing – Tweets with hashtags and trending content about Deepavali celebrations.

2. Holiday Announcements and Promotions – Posts mentioning the holiday period, discounts, and celebrations.

3. Community and Official Events – Tweets related to “Rumah Terbuka” (open house) events, “Madani” (government/community programs), and public gatherings.

G4. eneral Festive Greetings and Reflections – Broader discussions expressing well-wishes and cultural appreciation for Deepavali.

5. The Named Entity Recognition (NER) analysis detected entities such as Malaysia, India, October, and Deepavali, confirming the geographical and cultural relevance of the dataset. However, the NER model occasionally misclassified local or mixed-language phrases as PERSON or ORG, highlighting the challenge of processing multilingual Malaysian text.

Reflection

- This exercise demonstrates how NLP techniques can uncover patterns that might be missed through traditional manual analysis.

- Sentiment analysis quantifies public mood surrounding a cultural event.

- Topic modeling efficiently summarizes key discussion themes from hundreds of posts.

- NER helps identify important entities and recurring references, revealing what people associate with Deepavali.

However, there are limitations:

- The dataset size (200 tweets) is relatively small and may not represent all demographics.

- The multilingual nature of Malaysian tweets (mixing English, Malay, and Tamil) causes misclassifications and reduces accuracy for certain models like NER.

- Social media posts can be noisy containing emojis, links, and hashtags hich may affect word frequency and sentiment scores.

- Overall, this NLP analysis successfully illustrates that Deepavali in Malaysia is celebrated with positive sentiment, community spirit, and multicultural expression, reflecting the nation’s diverse and harmonious cultural landscape.
